# 00 - Create and register conda env

Creates the environment for the project 

```python
!mkdir -p env
```

```python
%%writefile ../env/environment.yml
name: gpu-serving
channels:
  - conda-forge
dependencies:
  - python=3.12.2
  - jupyter
  - ipykernel
  - numpy
  - requests
  - matplotlib
  - pip
```

```python
!echo $CONDA_ACCEPT_LICENSE
```

```python
%%bash
conda env create -f env/environment.yml -y
```

**Afterwards**:

1. activate it in your terminal: `conda activate gpu-serving`
2. restart IDE/notebook and select the `gpu-serving` kernel from the kernel menu
